```python
import scipy.special as ssp
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
```

```python
x = np.linspace(-4.0, 4.0, 40)
relu = np.copy(x)
relu[relu<0] = 0

lrelu = np.copy(x)
lrelu[lrelu<0] = 0.1 * lrelu[lrelu<0]

gelu = 0.5 * (1.0 + ssp.erf(x / np.sqrt(2.))) * x
# gelu2 = 0.5 * x * ( 1 + np.tanh(np.sqrt(2./np.pi) * x * (1. + 0.044715*x * x)))
gelu_min = np.min(gelu)

elu = np.copy(x[::2])
alpha = 1.0
elu[elu<0] = alpha * (np.exp(elu[elu<0]) - 1.)

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=gelu, name=f"gelu (sigma=1.0) and (min: {gelu_min:.2f})"))
# fig.add_trace(go.Scatter(x=x, y=gelu2, name="gelu2"))
fig.add_trace(go.Scatter(x=x, y=relu, name="relu"))
fig.add_trace(go.Scatter(x=x, y=lrelu, mode='markers', name="leaky relu"))
fig.add_trace(go.Scatter(x=x[::2], y=elu, mode='markers', name=f"elu (alpha={alpha:0.1f})"))
fig.show()
```
